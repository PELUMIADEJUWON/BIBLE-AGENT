```python
from dotenv import load_dotenv
```

```python
import os
```

```python
load_dotenv(override=True)
```

```python
from crewai import Agent, Task, Crew, LLM
```

```python
llm = LLM(
    model="openrouter/openai/gpt-4o-mini",
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("API_TOKEN")
)
```

```python
greeter = Agent(
    role="Friendly Greeter",
    goal="Greet people warmly and welcome them to bible study",
    backstory="You are the world's friendliest receptionist.",
    llm=llm,
    verbose=True
)
```

```python
greeting_task = Task(
    description="Tell the user Jesus loves them and welcome them to the christian fold. Give 3 bible passages to remind them of salvation.",  
    expected_output="A warm greeting followed by 3 benefits of salvation", 
    agent=greeter                                                        
)
```

```python
crew = Crew(
    agents=[greeter],           
    tasks=[greeting_task],   
    verbose=False                
)


result = crew.kickoff()         
print(result)
```
